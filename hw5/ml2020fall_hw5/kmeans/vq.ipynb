Run your k-means algorithm on `kmeans_data.mat`. Note that the lines indicate the iterations of centroids.

```python
from kmeans import kmeans
from kmeans_plot import kmeans_plot
x = sio.loadmat('kmeans_data.mat')['X']
k = 2  # modify if you want
#print(type(x))
max=0
min=1000
for i in range(1000):
    idxi, ctrsi, iter_ctrsi = kmeans(x, k)
    sd=0
    for label in np.unique(idxi):
        sd=np.sum(np.sqrt(np.sum(np.square(x[np.where(idxi==label)]-ctrsi[label]),axis=1)))+sd
    if sd>max:
        idx_max=idxi
        ctrs_max=ctrsi
        iter_ctrs_max=iter_ctrsi
        max=sd
    if sd<min:
        idx_min=idxi
        ctrs_min=ctrsi
        iter_ctrs_min=iter_ctrsi
        min=sd
#print(idx)
#print(type(ctrs))
#print(iter_ctrs)
kmeans_plot(x, idx_max, ctrs_max, iter_ctrs_max)
kmeans_plot(x, idx_min, ctrs_min, iter_ctrs_min)
```
